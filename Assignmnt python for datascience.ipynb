# Data Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Histogram
```

```python
oo.hist(bins=20, figsize=(10,8), grid=False)
plt.show()
```

```python
#  Bar plot for species count

plt.bar(oo['Classification'],oo['SL'],color='purple')
plt.xlabel('Species')
plt.ylabel('Sepal Length Count')
plt.title('Species Count Distribution Sepal Length')
plt.xticks(rotation=90)   
plt.show()
```

```python
plt.bar(oo['Classification'],oo['SW'],color='green')
plt.xlabel('Species')
plt.ylabel('SW Count')
plt.title('Species Count Distribution Sepal Width')
plt.xticks(rotation=90)   
plt.show()
```

```python
plt.bar(oo['Classification'],oo['PL'],color='pink')
plt.xlabel('Species')
plt.ylabel('PL Count')
plt.title('Species Count Distribution Petal Length')
plt.xticks(rotation=90)   
plt.show()
```

```python
plt.bar(oo['Classification'],oo['PW'],color='yellow')
plt.xlabel('Species')
plt.ylabel('PW Count')
plt.title('Species Count Distribution Petal Width')
plt.xticks(rotation=90)   
plt.show()
```

```python
species = oo['Classification']
plt.bar(species, oo['SL'], color='b', label='Sepal Length')
plt.bar(species, oo['SW'], bottom=oo['SL'], color='g', label='Sepal Width')
plt.bar(species, oo['PL'], bottom=oo['SL'] + oo['SW'], color='r', label='Petal Length')
plt.xlabel('Species')
plt.ylabel('Values')
plt.title('Stacked Bar Plot for Sepal Length, Sepal Width, and Petal Length')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Box plot for Variation of Sepal Length across different species
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(hue='Classification', y='SL',data=oo, palette='deep',legend=False)
plt.title('Variation of Sepal Length across different species')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.show()
```

```python
# Pair plot for relationships between features
```

```python
sns.pairplot(oo,hue='Classification',palette='Set1')
plt.show()
```
